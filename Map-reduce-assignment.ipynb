```python
import PyPDF2  # imported PyPDF2 for extracting data from the PDF

pdf_path = "C:/Users/lokes/Downloads/Harry_Potter_(www.ztcprep.com).pdf"

def extract_pages(pdf_path, start_page, end_page, output_filename):  # defined a function to extract text from specified pages
    with open(pdf_path, 'rb') as pdf_file: 
        pdf_reader = PyPDF2.PdfReader(pdf_file) 
        total_pages = len(pdf_reader.pages)  
        
        with open(output_filename, 'w', encoding='utf-8') as output_file:  # opening the output file to save extracted text
            for page_num in range(start_page - 1, end_page):  # looping through the specified range of pages
                if page_num < total_pages:  
                    page = pdf_reader.pages[page_num]  
                    text = page.extract_text()  
                    output_file.write(text)
                else:
                    print(f"Page {page_num + 1} is beyond the PDF length.")  # printing a message if page exceeds the total number of pages

# Extracting based on DOB: 23 January 1999
birth_day = 23 
birth_year = 1999 

# Extracting 10 pages starting from the page corresponding to birth day
start_day_page = birth_day  
extract_pages(pdf_path, start_day_page, start_day_page + 9, "birth_day_section.txt")  

# Extracting 10 pages starting from the page corresponding to birth year
start_year_page = birth_year % 100 
extract_pages(pdf_path, start_year_page, start_year_page + 9, "birth_year_section.txt")  

print("Text extraction completed.")  
```

```python
import re
from collections import defaultdict
from spellchecker import SpellChecker

# Function to read content from a file
def load_file_content(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return file.read()

# Map function to create a list of (word, 1) pairs from text
def create_word_pairs(text):
    words = re.findall(r'\b\w+\b', text.lower())  
    return [(word, 1) for word in words]  

# Reduce function to sum up word occurrences
def count_word_occurrences(word_pairs):
    occurrence_map = defaultdict(int)  
    for word, count in word_pairs:
        occurrence_map[word] += count
    return occurrence_map

# Function to identify non-English words
def extract_non_english_word_pairs(text, spell_checker):
    words = re.findall(r'\b\w+\b', text.lower()) 
    non_english_words = [word for word in words if word not in spell_checker] 
    return create_word_pairs(' '.join(non_english_words))  # Map non-English words to pairs

# Main function to perform word counting using MapReduce
def main():
    # Count occurrences of each word in source_file.txt
    source_file_path = 'birth_day_section.txt'
    source_text = load_file_content(source_file_path)
    
    word_pairs_source = create_word_pairs(source_text)  # Map step for source file
    word_occurrences_source = count_word_occurrences(word_pairs_source)  # Reduce step for source file
    
    print("Word counts from birth_day_section.txt:")
    for word, count in sorted(word_occurrences_source.items(), key=lambda x: x[1], reverse=True):
        print(f"{word}: {count}")

    # Count non-English words in reference_file.txt
    reference_file_path = 'birth_year_section.txt'
    reference_text = load_file_content(reference_file_path)
    
    spell_checker = SpellChecker()  # Initialize the spell checker
    non_english_word_pairs = extract_non_english_word_pairs(reference_text, spell_checker)  # Map step for non-English words
    non_english_word_occurrences = count_word_occurrences(non_english_word_pairs)  # Reduce step for non-English words
    
    print("\nNon-English word counts from birth_year_section.txt:")
    for word, count in sorted(non_english_word_occurrences.items(), key=lambda x: x[1], reverse=True):
        print(f"{word}: {count}")

if __name__ == "__main__":
    main()
```
